```python
import json
```

```python
with open("../data/word_annotations.json", 'r') as f:
    words = json.load(f)
```

```python
label_list = {
    "Brak etykiety": ["0"],
    "Obsługa": ["B-obsluga", "I-obsluga"],
    "Widoki": ["B-widoki", "I-widoki"],
    "Jedzenie": ["B-jedzenie", "I-jedzenie"],
    "Sen": ["B-sen", "I-sen"],
}

i = 0
label2id = {}
for label, value in label_list.items():
    label2id[label] = []
    for _ in value:
        label2id[label].append(i)
        i += 1
```

```python
label2id
```

```python
new_words = {}

for word in words:
    tokens = word["fragment"].split()
    if not tokens: # word["fragment"] == ""
        continue
    
    ids = label2id[word["label"]]
    ner_tags = [ids[0]] + [ids[-1]] * (len(tokens) - 1)

    assert len(tokens) == len(ner_tags)

    new_words.setdefault(
        word["text_id"], 
        {
            "id": word["text_id"],
            "ner_tags": [],
            "tokens": [],
            "full_text": ""
        }
    )
    new_words[word["text_id"]]["ner_tags"].extend(ner_tags)
    new_words[word["text_id"]]["tokens"].extend(tokens)
    new_words[word["text_id"]]["full_text"] += word["fragment"] + " "

tokens_annotations = list(new_words.values())

for text in tokens_annotations:
    assert len(text["ner_tags"]) == len(text["tokens"])
```

```python
with open("../data/token_annotations.json", 'w+') as f:
    json.dump(tokens_annotations, f)
```
